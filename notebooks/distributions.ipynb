```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from scipy.interpolate import interp1d
import json
```

# sigmoid

```python
c=-0.6
x0=12

x = np.linspace(0,20,1000)
y = 1/(1+np.exp(-c*(x-x0)))
plt.plot(x,y)
plt.xlim(0,20)
plt.ylim(0,1)
```

```python
r = x
fr_unnormed = y

norm = np.sum(2*np.pi*r*fr_unnormed)
fr_latent = 2*np.pi*r*fr_unnormed/norm

i = np.cumsum(fr_latent)
cdf_inv=interp1d(i,r)

u=np.random.uniform(size=(1000000,))
u = u[np.where((np.min(i)<u)&(np.max(i)>u))]
plt.hist(cdf_inv(u),bins=300,normed=True)
delta_r = r[1]-r[0]
plt.plot(r,fr_latent/delta_r,'r-')
plt.xlim(0,100)
```

```python
phi = 2*np.pi*np.random.uniform(size=np.size(u))

p = plt.hist2d(cdf_inv(u)*np.cos(phi)+20,cdf_inv(u)*np.sin(phi)+20,bins=40,cmap='Blues')
plt.axis('equal')
plt.colorbar()
```

```python
print(json.dumps({
    "request_origin_distributions":[{
        "x":list(x),
        "y":list(y),
        "location":[20,20]
}]}))
```

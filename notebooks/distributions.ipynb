# exponential

```python
from scipy.interpolate import interp1d
```

Surusegfuggveny.

```python
l=0.1

r = np.linspace(0,10000,100000)
fr_unnormed = l*np.exp(-l*r)
norm = np.sum(2*np.pi*r*fr_unnormed)
fr_latent = 2*np.pi*r*fr_unnormed/norm

print(np.sum(fr_latent))

plt.plot(r,fr_unnormed)
plt.xlim(0,20)
```

Eloszlasfuggveny.

```python
i = np.cumsum(fr_latent)
cdf_inv=interp1d(i,r)
plt.plot(r,i)
plt.plot(i,cdf_inv(i))
plt.xlim(0,1)
plt.ylim(0,1)
```

Pontok generalasa.

```python
u=np.random.uniform(size=(10000000,))
u = u[np.where((np.min(i)<u)&(np.max(i)>u))]
plt.hist(cdf_inv(u),bins=300,normed=True)
delta_r = r[1]-r[0]
plt.plot(r,fr_latent/delta_r,'r-')
plt.xlim(0,20)
```

Kiterites 2D-be.

```python
phi = 2*np.pi*np.random.uniform(size=np.size(u))
```

```python
p = plt.hist2d(cdf_inv(u)*np.cos(phi)+20,cdf_inv(u)*np.sin(phi)+20,bins=40,range=((0,40),(0,40)),cmap='Blues')
plt.xlim(0,40)
plt.ylim(0,40)
plt.axis('equal')
plt.colorbar()
```

# itt meg nem OK a normalas

```python
plt.plot(p[0][20,19:]/np.sum(p[0][20,19:]),label='cross-section')
plt.plot(r,fr_unnormed,label='original')
plt.xlim(0,20)
plt.legend()
```
